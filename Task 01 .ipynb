## **Linear Regression with Python Scikit Learn**
In this section we will see how the Python Scikit-Learn library for machine learning can be used to implement regression functions. We will start with simple linear regression involving two variables.

### **Simple Linear Regression**
In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

#### Step 01 : Importing libraries

```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import mean_squared_error , r2_score
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
```

#### Step 02 : Reading the data 

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")
data.head(5)
```

#### Step 03 : Plotting

Let's plot our data points on 2-D graph to eyeball our dataset and see if we can manually find any relationship between the data.

#### Plotting the Study Hours vs Scores Percentage 

```python
# scatter plot
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Study Hours vs Score Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
# scatter plot with a regression line 
sns.regplot(x="Hours" , y="Scores" , data=data)
plt.title('Study Hours vs Score Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
# bar plot 
data_sorted = data.sort_values(by='Hours')

# Create a bar plot
data_sorted.plot.bar(x='Hours', y='Scores')
plt.title('Study Hours vs Score Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

##### **From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.**

#### Step 04 : Prepering Data

Independent variable (attribute/input): Hours studied || 
Dependent variable (label/output): Score percentage

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

```python
print (X)
print (y)
```

#### Step 05 : Spit Data

Splitting the data into test and train data 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

#### Training the data with alogrithm

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

#### Testingg Data

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# predicitng the score based on hours
y_predic = regressor.predict([[9.5]])
y_predic
```

```python
print (y_pred , y_test , y_train)
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual Score': y_test,  'Predicted': y_pred })  
df 
```

```python
sns.lineplot(data = df)
plt.title("Actual vs Predicted Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
```

#### Step 06 : Evaluating

 Evaluate the performance of algorithm

```python
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
print ('R2 Scroe for linear Regression' , r2_score(y_test , y_pred))
```
